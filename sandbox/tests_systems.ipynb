```python
import nglview as nv
import healpy as hp
import numpy as np
from scipy.spatial.distance import euclidean
from scipy.spatial import Voronoi
from math import sqrt
```

```python
nside = 4

n_points=hp.nside2npix(nside)
points = np.zeros((n_points, 3))
x, y, z = hp.pix2vec(nside,np.arange(n_points),nest=False)
points[:,0]=x
points[:,1]=y
points[:,2]=z
```

```python
voronoi = Voronoi(points)
```

```python
alphaspheres = voronoi.vertices
n_alphaspheres = alphaspheres.shape[0]
```

```python
points_of_alphasphere = {ii:[] for ii in range(n_alphaspheres)}
points_of_alphasphere[-1]=[]

face={}
```

```python
for point_index, region_index in enumerate(voronoi.point_region):
    for as_index in voronoi.regions[region_index]:
        points_of_alphasphere[as_index].append(point_index)
```
